# Q2-In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a dataframe and csv. In case if any product vertical has
less than 3 pages in search results then scrape all the products available under that product
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”. In case, if any of the details are missing for any of the product then
replace it by “-“

```python
url=[]
start=1
end=3
for page in range(start,end):
    url_tags=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for i in url_tags:
        url.append(i.get_attribute('href'))
    
    next_button=driver.find_element_by_xpath("//li[@class='a-last']/a")
    nxt=next_button.get_attribute('href')
    driver.get(nxt)
```

```python
brand=[]
product=[]
rating=[]
total_rating=[]
availability=[]
price=[]
delivery=[]
replacement=[]
other=[]
URL=[]

for ur in url:
    driver.get(ur)
    try:
        brand_tag=driver.find_element_by_xpath("//a[@id='bylineInfo']")
        brand.append(brand_tag.text)
    except NoSuchElementException:
        brand.append("--")
        
    try:
        product_tag=driver.find_element_by_xpath("//h1[@class='a-size-large a-spacing-none']/span")
        product.append(product_tag.text)
    except NoSuchElementException:
        product.append('--')
        
    try:
        rating_tag=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        rating.append(rating_tag.text)
    except NoSuchElementException:
        rating.append('--')
        
    try:
        total_tag=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        total_rating.append(total_tag.text)
    except NoSuchElementException:
        total_rating.append('--')
        
    try:    
        availability_tag=driver.find_element_by_xpath("//div[@id='availability']/span")
        availability.append(availability_tag.text)
    except NoSuchElementException:
        availability.append('--')
        
    try:
        price_tag=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']")
        price.append(price_tag.text)
    except NoSuchElementException:
        price.append('--')
        
    try:
        delivery_tag=driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b")
        delivery.append(delivery_tag.text)
    except NoSuchElementException:
        delivery.append('--')
        
    try:
        replacement_tag=driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
        replacement.append(replacement_tag.text)
    except NoSuchElementException:
        replacement.append('--')
        
    try: 
        other_tag=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']")
        other.append(other_tag.text)
    except NoSuchElementException:
        other.append('--')
        
    URL.append(ur) 
    time.sleep(5)
```

```python
df=pd.DataFrame({'Brand':brand,'Product':product,'Price':price,'Rating':rating,'Total_Rating':total_rating,'Availability':availability,'Delivery':delivery,'Replacement':replacement,'Other_detail':other,'URL':URL})
```
